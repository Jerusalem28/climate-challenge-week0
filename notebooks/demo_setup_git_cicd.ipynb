## Section 1 — Installing Python on Your Machine

Before you can create virtual environments or run any Python code, you need **Python itself** installed on your system.

### How to check if Python is already installed

Open a terminal and run:
- **macOS/Linux**: `python3 --version`
- **Windows**: `python --version` or `py --version`

If you see a version number (e.g., `Python 3.11.9`), Python is installed. If you get "command not found", you need to install it.

### Installing Python — by Operating System

| OS | Recommended Method | Steps |
|----|-------------------|-------|
| **macOS** | Homebrew (recommended) | `brew install python` |
| **macOS** | python.org installer | Download from [python.org/downloads](https://python.org/downloads) |
| **Windows** | Microsoft Store | Search "Python" in the Store, install Python 3.11+ |
| **Windows** | python.org installer | Download the `.exe` — **Check "Add Python to PATH"** |
| **Linux (Ubuntu/Debian)** | apt | `sudo apt update && sudo apt install python3 python3-pip python3-venv` |
| **Linux (Fedora/RHEL)** | dnf | `sudo dnf install python3 python3-pip` |

## Section 1.3 — Conda: What It Is and How to Set It Up

### What is Conda?

**Conda** is an open-source **package manager AND environment manager** — it does the job of both `pip` and `venv` in one tool.

> **When to use conda:** data science / ML projects that rely on heavy libraries like NumPy, PyTorch, TensorFlow, or anything that needs compiled C/C++ extensions. It resolves binary dependencies that `pip` sometimes struggles with.

````python
# ─────────────────────────────────────────────────────────────────────────────
#  1.3  Conda setup — install, create, activate, manage
# ─────────────────────────────────────────────────────────────────────────────
#
#  ── STEP 1: Install Miniconda ────────────────────────────────────────────────
#  Download the installer from: https://docs.conda.io/en/latest/miniconda.html
#
#  macOS / Linux (run in terminal):
#    bash Miniconda3-latest-MacOSX-x86_64.sh    # Intel Mac
#    bash Miniconda3-latest-MacOSX-arm64.sh     # Apple Silicon (M1/M2/M3)
#    bash Miniconda3-latest-Linux-x86_64.sh     # Linux
#
#  Windows:
#    Double-click the .exe installer and follow the prompts.
#
#  After installation, RESTART your terminal, then verify:
#    conda --version
#
# ─────────────────────────────────────────────────────────────────────────────
#
#  ── STEP 2: Create a new environment ─────────────────────────────────────────
#
#  conda create -n myenv python=3.11
#    │             │         └─ pin the Python version for this env
#    │             └─ name of the environment (you choose this)
#    └─ create a new conda environment
#
#  conda create -n myenv python=3.11 pandas numpy   # install packages at creation time
#
# ─────────────────────────────────────────────────────────────────────────────
#
#  ── STEP 3: Activate the environment ─────────────────────────────────────────
#
#  conda activate myenv        # activate (prompt changes to: (myenv) $)
#  conda deactivate            # exit the environment (go back to base)
#
# ─────────────────────────────────────────────────────────────────────────────
#
#  ── STEP 4: Install packages inside the environment ──────────────────────────
#
#  conda install numpy pandas matplotlib    # install from the conda channel
#  conda install -c conda-forge seaborn     # install from a community channel
#  pip install some-package                 # you can still use pip inside conda!
#
# ─────────────────────────────────────────────────────────────────────────────
#
#  ── STEP 5: Useful conda commands ────────────────────────────────────────────
#
#  conda env list                  # list all environments (* = active)
#  conda list                      # list all packages in the current env
#  conda info                      # show current conda config and active env
#  conda remove -n myenv --all     # delete an environment completely
#
# ─────────────────────────────────────────────────────────────────────────────
````

## Section 3 — `.gitignore`

.gitignore — tell Git what NOT to track

- .venv/ → virtual environment (huge, OS-specific)
- .env  → secrets & API keys — NEVER commit!
- data/  → raw data files (challenge requirement)
- *.csv  → data files
- __pycache__/  → Python bytecode (auto-generated)


````python
# ─────────────────────────────────────────────────────────────────────────────
#  3.1  Read & display the project .gitignore
# ─────────────────────────────────────────────────────────────────────────────

import pathlib

# The .gitignore lives one level up from this notebook
gi_path = pathlib.Path('..') / '.gitignore'

if gi_path.exists():
    print(gi_path.read_text())
else:
    print('⚠️  .gitignore not found — create one in the project root!')
````

## Section 5: Git fundamentals


````python
# ─────────────────────────────────────────────────────────────────────────────
# git init — turn a folder into a Git repository

# ─────────────────────────────────────────────────────────────────────────────
#
# cd path/to/your/project     # navigate to your project folder
# git init                    # initialise — creates a hidden .git/ folder
#
# ── What git init creates ────────────────────────────────────────────────────
#   .git/          ← the entire history database lives here
#     HEAD         ← which branch you are currently on
#     config       ← repo-level git settings
#     objects/     ← compressed snapshots of every commit
#     refs/        ← pointers to branch tips
#
# ⚠️  Never delete or edit .git/ manually — you will lose all history.
# ─────────────────────────────────────────────────────────────────────────────
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# git status
#
# Check what git would currently track (before committing)
# ─────────────────────────────────────────────────────────────────────────────
#
# git status
#    ← red  = untracked files (git sees them but is NOT tracking them yet)
#    ← green = staged files (ready to be committed)
# ─────────────────────────────────────────────────────────────────────────────
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# git add — move files from Working Directory → Staging Area
# ─────────────────────────────────────────────────────────────────────────────
#
# git add .                   # stage ALL changed files at once
# git add README.md           # stage ONE specific file
# git add src/ tests/         # stage an entire folder
````

````python
# ─────────────────────────────────────────────────────────────────────────────
#  git commit — save the staged snapshot permanently
#
# git commit -m "init: add .gitignore, README and requirements.txt"
#
# ── Examples ─────────────────────────────────────────────────────────────────
#   git commit -m "init: add .gitignore"
#   git commit -m "chore: venv setup"
#   git commit -m "ci: add GitHub Actions workflow"
#   git commit -m "feat(eda): add temperature trend plot for Ethiopia"
# ─────────────────────────────────────────────────────────────────────────────
````

## Section 5 — GitHub Workflow

| Term | Meaning |
|------|---------|
| `origin` | The remote copy of your repo on GitHub |
| `git clone` | Download a repo from GitHub to your machine |
| `git push` | Upload your local commits to GitHub |
| `git pull` | Download + merge GitHub changes into your local copy |
| Pull Request | Ask for your branch to be reviewed and merged into `main` |

````python
# ─────────────────────────────────────────────────────────────────────────────
#  Key remote commands reference
#
# ── Link a local repo to GitHub (if you did git init locally) ────────────────
# git remote add origin https://github.com/<user>/repo.git
# git push -u origin main          # first push — sets the tracking branch
#
# ── Everyday push (after the first) ──────────────────────────────────────────
# git push
#
# ── Push a feature branch ────────────────────────────────────────────────────
# git push origin setup-task
#
# ── Get the latest from GitHub (teammates' commits) ──────────────────────────
# git pull origin main
````

### git clone — Download a Repository FROM GitHub TO Your Machine

`git clone` creates a local copy of a remote repository. Use this when:
- You want to contribute to an existing project
- You're starting work on a different machine
- You're joining a team and need their codebase

````python
# ─────────────────────────────────────────────────────────────────────────────
#  git clone — copy a GitHub repo to your local machine
# ─────────────────────────────────────────────────────────────────────────────

# STEP 1: Get the repository URL from GitHub
#   - Go to the GitHub repo page
#   - Click the green "<> Code" button
#   - Copy the HTTPS URL: https://github.com/username/repo-name.git

# STEP 2: Clone the repository
#
# git clone https://github.com/username/repo-name.git
#    │       └─ the URL you copied from GitHub
#    └─ creates a new folder with the repo name and downloads all files + history
#
# This creates a folder called "repo-name" in your current directory with:
#   - All source files
#   - The entire git history (all commits, branches)
#   - The remote already configured as "origin"


# ── git clone vs git init ─────────────────────────────────────────────────────
#
# ┌─────────────────┬──────────────────────────────┬──────────────────────────────┐
# │                 │ git init                     │ git clone                  │
# ├─────────────────┼──────────────────────────────┼──────────────────────────────┤
# │ When to use     │ Starting a NEW project       │ Working on EXISTING repo    │
# │ Creates         │ Empty .git/ folder           │ Full copy of remote repo   │
# │ Remote setup    │ You must add:                │ Already configured as      │
# │                 │ git remote add origin <url>  │ "origin"                   │
# │ First commit    │ You create the first commit  │ History already exists     │
# └─────────────────┴──────────────────────────────┴──────────────────────────────┘
#
# ─────────────────────────────────────────────────────────────────────────────
````

## Section 6 — CI/CD with GitHub Actions

**CI (Continuous Integration)** = every `git push` automatically runs install + lint + tests.  
**CD (Continuous Deployment)** = if CI passes, automatically deploy the app.

```
git push
   │
   ▼
GitHub Actions triggered
   │
   ├─► 1. Checkout code
   ├─► 2. Set up Python 3.11
   ├─► 3. pip install -r requirements.txt
   ├─► 4. flake8  (lint)
   └─► 5. pytest  (tests)
              │
        ✅ PASS → safe to merge PR
        ❌ FAIL → fix before merging
```

The workflow file lives at: **`.github/workflows/ci.yml`**

````python
# ─────────────────────────────────────────────────────────────────────────────
# Add CI workflow file
#
# mkdir -p .github/workflows
# # ... create .github/workflows/ci.yml
# git add .github/
# git commit -m "ci: add GitHub Actions workflow"
# ─────────────────────────────────────────────────────────────────────────────


ci_content = '''name: CI Pipeline

on:
  push:
    branches: [main]
  pull_request:
    branches: [main]

jobs:
  build-and-test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: actions/setup-python@v5
        with:
          python-version: "3.11"
      - run: python --version
      - run: pip install -r requirements.txt
      - run: flake8 src/ tests/ --max-line-length=88
      - run: pytest tests/ -v --tb=short
'''
````
